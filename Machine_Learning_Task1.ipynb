```python
# Importing the libraries
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
print("Setup Complete")
```

```python
# Importing the dataset
filepath = "C:/Users/wanee/Desktop/Technohack_Internship_Tasks/Machine_Learning/iris.csv"
file = pd.read_csv(filepath)
file
```

```python
#Iris-setosa = 0
#Iris-versicolor = 1
#Iris-virginica = 2
file['Flower'] = file['Flower'].replace({'Iris-setosa':0, 'Iris-versicolor':1, 'Iris-virginica':2})
file
```

```python
X = file.drop(['Flower'], axis=1)
y = file['Flower']
print(X)
print(y)
```

```python
# Checking if the data needs to be cleaned
file.isnull().sum()
```

```python
# Data Analysis
file.describe()
```

```python
# Test/Train Split
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state = 0)
# Define model
model = DecisionTreeRegressor(random_state=0)
# Fit model
model.fit(train_X, train_y)

# get predicted prices on validation data
val_predictions = model.predict(val_X)
acc = accuracy_score(val_y, val_predictions)*100
print(f'Accuracy is {acc: .2f}%')
```
